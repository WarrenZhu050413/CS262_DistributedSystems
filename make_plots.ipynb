```python
import pandas as pd
#read in json_efficiency.csv, and custom_efficiency.csv
json_efficiency = pd.read_csv('json_efficiency.csv')
custom_efficiency = pd.read_csv('custom_efficiency.csv')

# filter both dataframes to include only column Operation = 'encode'
json_encode = json_efficiency[json_efficiency['Operation'] == 'encode']
custom_encode = custom_efficiency[custom_efficiency['Operation'] == 'encode']

# plot the average values of the column 'Elapsed_Time' for each dataframe, using a bar plot for both
json_encode['Elapsed_Time'].mean()
custom_encode['Elapsed_Time'].mean()
import matplotlib.pyplot as plt
plt.bar(['json', 'custom'], [json_encode['Elapsed_Time'].mean(), custom_encode['Elapsed_Time'].mean()])
plt.ylabel('Elapsed Time (s)')
plt.title('Average Efficiency')
plt.show()
```
